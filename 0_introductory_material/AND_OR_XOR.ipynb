# **Boolean Συναρτησεις**

*1) AND*

*2) OR*

*3) XOR*

```python
import numpy as np
import matplotlib.pyplot as plt

#Δυαδικές είσοδοι
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
```

```python
#Λογικές συναρτήσεις
y_and = np.logical_and(X[:,0], X[:,1]).astype(int)
y_or  = np.logical_or(X[:,0], X[:,1]).astype(int)
y_xor = np.logical_xor(X[:,0], X[:,1]).astype(int)
```

```python
def plot_logic_with_line(ax, X, y, title, w, b):

    #Να μην μπαίνουν πολλαπλά ίδια στο label
    label_1_added = False
    label_0_added = False

    for i in range(len(y)):
        if y[i] == 1:
            if (label_1_added == False):
              ax.scatter(X[i,0], X[i,1], marker='o', color='r', label='Έξοδος 1')
              label_1_added = True
            else:
              ax.scatter(X[i,0], X[i,1], marker='o', color='r')
        else:
            if (label_0_added == False):
              ax.scatter(X[i,0], X[i,1], marker='x', color='b', label='Έξοδος 0')
              label_0_added = True
            else:
              ax.scatter(X[i,0], X[i,1], marker='x', color='b')

    #Αν δοθούν βάρη, σχεδίασε ευθεία απόφασης
    if w is not None and b is not None:
        x_vals = np.linspace(-0.1, 1.1, 100)
        y_vals = -(w[0]*x_vals + b) / w[1]
        ax.plot(x_vals, y_vals, label='Ευθεία Διαχωρισμού Κατηγοριών', color='purple')

    ax.legend(loc='center').get_frame().set_edgecolor('black')
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_title(title)
    ax.grid(True)
    ax.set_xlim(-0.1, 1.1)
    ax.set_ylim(-0.1, 1.1)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

title_and = 'AND'
title_or  = 'OR'
title_xor = 'XOR'

plot_logic_with_line(axes[0], X, y_and, title_and, w=[1, 1], b=-1.5)
plot_logic_with_line(axes[1], X, y_or,  title_or, w=[1, 1], b=-0.5)
plot_logic_with_line(axes[2], X, y_xor, title_xor, w=None, b=None)

plt.tight_layout()
plt.show()
```
