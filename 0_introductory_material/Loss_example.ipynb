```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
w = np.linspace(-2, 2, 100)
b = np.linspace(-2, 2, 100)
W, B = np.meshgrid(w, b)

#  J(w, b)
J = W**2 + B**2

#
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

#Surface plot
surf = ax.plot_surface(W, B, J, cmap='hot')

ax.set_xlabel('W')
ax.set_ylabel('b')
ax.set_zlabel('J(w,b)')
#plt.title('Surface plot    J(w,b)')
plt.show()
```
