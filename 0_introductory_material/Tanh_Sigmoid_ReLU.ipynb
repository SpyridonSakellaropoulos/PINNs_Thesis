# **Γραφήματα των Συναρτήσεις Ενεργοποίησης**

### **1) Υπερβολική Εφαπτομένη**

$$
Tanh(z) = \frac{ e^{z} - e^{-z} }{ e^z + e^{-z} }
$$

### **2) Σιγμοειδής**

$$
Sigmoid(z) = \frac{1}{ 1 + e^{-z} }
$$

### **3) ReLU (Rectified Linear Unit)**

$$
ReLU(z) = max(0,z)
$$

### **4) Softplus**

$$
Softplus(z) = \ln(1 + e^z)
$$

### **5)Swish**

$$
Swish(z) = z\ Sigmoid(\beta\ z) = \frac{z}{1 + e^{- \beta\ z}}
$$



```python
import torch                    #βασικό πακέτο PyTorch
import matplotlib.pyplot as plt #βιβλιοθήκη γραφήματα
```

```python
#Υπερβολική Εφαπτομένη
def tanh(z):
  return (torch.exp(z) - torch.exp(-z))/(torch.exp(z) + torch.exp(-z))

#Σιγμοειδής
def sigmoid(z):
  return 1.0/(1.0 + torch.exp(-z))

#ReLU
def relu(z):
  return torch.maximum(torch.zeros_like(z),z)

#Softplus
def softplus(z):
  return torch.log(1 + torch.exp(z))

#Swish
def swish(z,b):
  return z*sigmoid(b*z)
```

```python
#αριθμός σημείων
N = 1000

#σημεία
z = torch.linspace(-10,10,N)

#αποτελέσματα συναρτήσεων
y_tanh = tanh(z)
y_sigmoid = sigmoid(z)
y_relu = relu(z)
y_softplus = softplus(z)
y_swish = swish(z, b=1)
```

```python
#γραφήματα
fig = plt.figure(figsize=(6,5))

porfiro_klasiko = '#800080'
porfiro_skoyro = '#6a0dad'
```

```python
plt.plot(z.numpy(), y_tanh.numpy(), color=porfiro_skoyro, linewidth=2.5, label='y=tanh(z)')
plt.xlabel('$z$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```

```python
plt.plot(z.numpy(), y_sigmoid.numpy(), color=porfiro_skoyro, linewidth=2.5, label='y=sigmoid(z)')
plt.xlabel('$z$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```

```python
plt.plot(z.numpy(), y_relu.numpy(), color=porfiro_skoyro, linewidth=2.5, label='y=relu(z)')
plt.xlabel('$z$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```

```python
plt.plot(z.numpy(), y_softplus.numpy(), color=porfiro_skoyro, linewidth=2.5, label='y=Softplus(z)')
plt.xlabel('$z$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```

```python
plt.plot(z.numpy(), y_swish.numpy(), color=porfiro_skoyro, linewidth=2.5, label='y=Swish(z)')
plt.xlabel('$z$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```
